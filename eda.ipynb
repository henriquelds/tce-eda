# EDA em dados de licitação do TCE-RS

###### Carregando bibliotecas úteis..

```python
import os,sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
%matplotlib inline
```

###### Setando o ano e carregando dataset licitacao correspondente

```python
year = 2016
path_lic = os.path.join(".",str(year),"licitacao.csv")
df_lic = pd.read_csv(path_lic)
```

###### Removendo licitacoes com tipo diferente de COMPRA, colunas com mais do que 50% de NaN, colunas sem significado útil e observações com VL_LICITACAO == NaN.

```python
df_lic.drop(df_lic[df_lic['TP_OBJETO'] != "COM"].index, inplace=True)
df_lic = df_lic.loc[:, df_lic.isnull().sum() < 0.5*df_lic.shape[0]]
df_lic.drop(["NR_COMISSAO", "ANO_COMISSAO", "NR_PROCESSO", "ANO_PROCESSO",
             "CD_TIPO_FASE_ATUAL", "TP_NIVEL_JULGAMENTO", "TP_CARACTERISTICA_OBJETO",
             "TP_NATUREZA", "TP_OBJETO", "BL_PERMITE_SUBCONTRATACAO", "BL_INVERSAO_FASES",
             "DT_HOMOLOGACAO", "DT_ADJUDICACAO", "BL_LICIT_PROPRIA_ORGAO", "BL_GERA_DESPESA",
             "BL_COMPARTILHADA"], axis=1, inplace=True)
df_lic.dropna(subset=['VL_LICITACAO'], inplace=True)
df_lic.shape
```

###### Algumas variáveis categóricas remanescentes tem distribuições bem desiguais, vamos agrupar esses valores em categorias denomiadas OUTRO. Cada coluna terá um threshold definido arbitrariamente, decidido por mim após inspeção visual (idealmente poderia ser feito por percentil).

```python
df_lic = df_lic.apply(lambda c: c.mask(c.map(c.value_counts()) < 400, "OUTRO") if c.name=="CD_TIPO_MODALIDADE" else c )
df_lic = df_lic.apply(lambda c: c.mask(c.map(c.value_counts()) < 400, "OUTRO") if c.name=="TP_COMISSAO" else c )
df_lic = df_lic.apply(lambda c: c.mask(c.map(c.value_counts()) < 400, "OUTRO") if c.name=="TP_LICITACAO" else c )
df_lic = df_lic.apply(lambda c: c.mask(c.map(c.value_counts()) < 400, "OUTRO") if c.name=="TP_BENEFICIO_MICRO_EPP" else c )
```

###### Analisando como o valor médio/soma de licitação varia ao longo do ano: criamos uma coluna com o mês da abertura da licitação, agrupamos por mês e plotamos 2 histogramas simples.

```python
df_lic["MES_ABERTURA"] = pd.DatetimeIndex(df_lic["DT_ABERTURA"]).month
df_aux = df_lic[["VL_LICITACAO", "MES_ABERTURA"]]
df_aux = df_aux.groupby(['MES_ABERTURA'], as_index=False).agg({'VL_LICITACAO': ['mean', 'sum']})
df_aux = df_aux.xs('VL_LICITACAO', axis=1, drop_level=True).rename_axis('Mês').reset_index()
df_aux["Mês"] += 1
ax = sns.catplot(x='Mês', y='mean', kind="bar", data=df_aux)
ax.fig.suptitle("Valor médio de licitação por mês")
plt.show()
```

```python
ax = sns.catplot(x='Mês', y='sum', kind="bar", data=df_aux)
ax.fig.suptitle("Valor total de licitação por mês")
plt.show()
```

######  Julho e Dezembro são os meses com maior acúmulo no valor de licitações. Vamos ver quais orgãos acumulam maiores valores de licitações no ano todo e a cada mês:
Para isso vamos ter que aumentar a granularidade da análise. São cerca de 600 orgãos únicos, qualquer visualização ou análise nesse domínio fica pouco explicativa. Assim, vamos agrupá-los em 7 grupos: prefeituras, câmaras municipais, conselhos, bancos, fundações, empresas (CIAs) e autarquias

```python
df_lic["TP_ORGAO"] = np.where( df_lic['NM_ORGAO'].str.contains("PM DE "), "Prefeituras",
                     np.where( df_lic['NM_ORGAO'].str.contains("CM DE "), "Câmaras",
                     np.where( df_lic['NM_ORGAO'].str.contains("CONS. "), "Conselhos",
                     np.where( df_lic['NM_ORGAO'].str.contains("BANCO "), "Bancos",
                     np.where( df_lic['NM_ORGAO'].str.contains("FUND."), "Fundações",
                     np.where( df_lic['NM_ORGAO'].str.contains("FUNDAÇÃO"), "Fundações",
                     np.where( df_lic['NM_ORGAO'].str.contains("CIA."), "Empresas", 
                     np.where( df_lic['NM_ORGAO'].str.contains("EMPRESA"), "Empresas", "Autarquias"        
                     ))))))))
```

###### Quando consideramos o ano todo:
podemos plotar um boxplot de cada tipo de orgão

```python
df_lic.boxplot(column="VL_LICITACAO", by="TP_ORGAO", showfliers=False)
plt.ylabel("VL_LICITACAO")
plt.suptitle('')
plt.title('')
plt.show()
```

###### O boxplot é pouco conclusivo:
limites superiores muito distantes porém, aparentemente, a mediana de todos os tipos de orgãos está no mesmo nível. O que se pode dizer é que câmaras e prefeituras possuem a menor variância entre valores de licitação, e conselhos a maior variância. Vamos rodar um teste não-paramétrico para cada combinação de valores de licitação por órgao e verificar se existe diferença estatística significativa entre diferentes tipos de orgãos.

```python
from itertools import combinations
from scipy.stats import mannwhitneyu
ids = list(df_lic["TP_ORGAO"].unique())
mat = np.ones((len(ids),len(ids)))
for i in list(combinations(ids,2)):
    dist0 = list(df_lic[df_lic['TP_ORGAO'] == i[0]]["VL_LICITACAO"])
    dist1 = list(df_lic[df_lic['TP_ORGAO'] == i[1]]["VL_LICITACAO"])
    _, p_value = mannwhitneyu(dist0,dist1)
    mat[ ids.index(i[0]), ids.index(i[1])] = p_value
df_aux = pd.DataFrame(mat, index=ids, columns=ids)
ax = sns.heatmap(df_aux, cmap='RdYlGn_r', linewidth=0.2, annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom+0.5, top-0.5)
```

###### Conclusão (desconsiderar células com 1, é apenas uma matriz diagonal superior):
Não há diferença significativa entre os valores de licitações de prefeituras e fundações, de autarquias e empresas, de autarquias e conselhos, conselhos e empresas, fundações e conselhos.
No entanto, deve-se considerar que há uma disparidade muito grande na quantidade de licitações anual por tipo de orgão, como visto abaixo:

###### Aproximar o valor da licitação com modelo linear:
Consideramos tanto mês quanto tipo de orgão como variáveis categóricas.
- Primeiro, utilizaremos o dataset inteiro pré-processado até agora, ou seja, observações individuais de VL_LICITACAO com variáveis categóricas originais + TP_ORGAO + MES_ABERTURA.
- Segundo, vamos acrescentar features numéricas referentes a descrição da licitação (DS_OBJETO) -- vetor TF-IDF e tentaremos aproximar o valor individual de licitação
- Terceiro, vamos acrescentar features numéricas do texto da Licitação + items a partir de um modelo neural doc2vec.

Nesse experimento, o objetivo é verificar qual variável contribui mais para uma aproximação linear de VL_LICITACAO. Vamos utilizar todas as variáveis preditoras disponíveis, sendo que todas elas são categóricas, inclusive o mês. Para isso, vamos utilizar o método OLS - ordinary least squares -- que parametriza uma função linear das variáveis preditoras minimizando a soma dos quadrados das diferenças entre a variável preditiva (VL_LICITACAO) e o seu valor predito.

Além disso, pensando num produto que seria útil para servidores que fiscalizam licitações, se temos um modelo treinado com variáveis de entrada que normalmente produz um bom valor predito de VL_LICITACAO, seria possível utilizá-lo em teste para detectar licitações com valores exorbitantes. Isto é, para uma dada entrada/licitação, se o valor da licitação está muito acima do valor que um modelo bem treinado/aproximado prediz, então pode haver indícios de fraude.

###### Mas antes...
Vamos converter as variáveis categóricas em dummy (dropando a primeira categoria pra evitar dummy trap em regressão linear) e verificar as correlações

###### Primeiro modelo:

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

X = df_aux_dum.drop(["VL_LICITACAO"], axis=1, inplace=False)
y = df_aux_dum[["VL_LICITACAO"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
model = sm.OLS(y_train,X_train)
fitted = model.fit()
fitted.summary()
```

###### Sobre as estatísticas:
Valor muito baixo de R-squared indica que o modelo não conseguiu "fitar" corretamente a curva de treino. Na verdade o desempenho em treino foi tão baixo que nem vale a pena testar o modelo nas observações de teste.
É possível ver, no entanto, que as variáveis com maior potencial preditivo (de acordo com p-value) são BL_ORCAMENTO_SIGILOSO_S e TP_FORNECIMENTO_P, ambas com coeficiente positivo, indicando que orçamentos sigilosos e com fornecimento parcelado tendem a custar mais (o que não deixa de ser um pouco óbvio...). Também TP_ORGAO_Conselhos parece ser uma variável útil para estimar o VL_LICITACAO.

###### Hipótese:
É possível que valores outliers de VL_LICITACAO estejam degradando o modelo. Vamos removê-los usando a técnica do intervalo interquartil e re-treinar o modelo.

```python
q_low = df_aux_dum["VL_LICITACAO"].quantile(0.10)
q_high = df_aux_dum["VL_LICITACAO"].quantile(0.90)
df_aux_dum = df_aux_dum[(df_aux_dum["VL_LICITACAO"] < q_high) & (df_aux_dum["VL_LICITACAO"] > q_low)]
```

```python
X = df_aux_dum.drop(["VL_LICITACAO"], axis=1, inplace=False)
y = df_aux_dum[["VL_LICITACAO"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
model = sm.OLS(y_train,X_train)
fitted = model.fit()
fitted.summary()
```

###### A hipótese estava certa!
Conseguimos um R-squared de 0.48 (varia de 0 a 1). Não é um modelo ótimo mas já indica que é possível aproximar VL_LICITACAO linearmente. Também se vê que as variáveis mais importantes são outras:

- MES_ABERTURA_3/4/11/12 - também estão ligados a valores mais altos de licitação
Vamos verificar no teste:

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
predictions = fitted.predict(X_test)
labels = np.array(list(y_test["VL_LICITACAO"]))
preds = np.array(predictions.tolist())
rmse = sqrt(mean_squared_error(labels,preds))
rmse
```

###### Vamos para o segundo modelo:
Vamos adicionar variáveis numéricas baseadas no modelo de contagem de palavras TF-IDF da descrição da licitação DS_OBJETO. Cada documento (licitação) se torna um vetor de tamanho N, onde N é número total de palavras do campo DS_OBJETO. Em cada documento, na posição i do vetor é colocado um valor tf(wi).idf(wi) onde tf(wi) é a frequência da palavra wi naquele documento, e idf(wi) é o inverso da frequência da palavra no corpus inteiro. Assim, keywords recebem peso maior ao passo que palavras que repetem muito ao longo do corpus recebem peso menor. Além disso, iremos eliminar stopwords da lingua portuguesa.

###### Antes vamos recriar nosso dataset dummificado mas dessa vez sem excluir a coluna DS_OBJETO:
também vamos remover os outliers e a coluna TP_LICITACAO_NSA, como fizemos antes.

```python
df_aux = df_lic.drop(["NM_ORGAO", "CD_ORGAO","NR_LICITACAO", "ANO_LICITACAO", "DT_ABERTURA"], axis=1, inplace=False )
df_aux_dum = pd.concat((df_aux[["VL_LICITACAO", "DS_OBJETO"]],
                      pd.get_dummies(data=df_aux, columns=["MES_ABERTURA","CD_TIPO_MODALIDADE",
                                    "TP_COMISSAO","TP_LICITACAO", "TP_ORGAO",
                                    "TP_BENEFICIO_MICRO_EPP","TP_FORNECIMENTO",
                                    "BL_ORCAMENTO_SIGILOSO", "BL_PERMITE_CONSORCIO"], drop_first=True)), axis=1)
df_aux_dum = df_aux_dum.loc[:,~df_aux_dum.columns.duplicated()]
q_low = df_aux_dum["VL_LICITACAO"].quantile(0.10)
q_high = df_aux_dum["VL_LICITACAO"].quantile(0.90)
df_aux_dum = df_aux_dum[(df_aux_dum["VL_LICITACAO"] < q_high) & (df_aux_dum["VL_LICITACAO"] > q_low)]
df_aux_dum.drop(["TP_LICITACAO_NSA"], axis=1, inplace=True)
df_aux_dum.reset_index(inplace=True,drop=True)
```

###### Agora podemos computar o vetor tf-idf de cada licitação
dessa vez vamos usar também um HashingVectorizer, que normaliza o vetor tf.

```python
from sklearn.feature_extraction.text import TfidfTransformer,HashingVectorizer
from sklearn.pipeline import make_pipeline
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
hasher = HashingVectorizer(n_features=10000, stop_words=stopwords.words('portuguese')) #max 10000 features
v = make_pipeline(hasher, TfidfTransformer())
lic_embed = v.fit_transform(df_aux_dum["DS_OBJETO"])
```

###### São 10k features.. obviamente não vamos usar 10k variáveis, podemos reduzir os componentes desse vetor que representa o texto da licitacao usando SVD (garantindo ortogonalidade nas dimensões resultantes):
logo após, concatenamos as variáveis ao nosso dataset.

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=5) #Escolhemos manter apenas 5 dimensões/variáveis..
lic_embed_red = svd.fit_transform(lic_embed)
df_pcs = pd.DataFrame(lic_embed_red, dtype=np.float32, columns=["p1", "p2", "p3", "p4", "p5"])
df_aux = pd.concat([df_aux_dum,  df_pcs], axis=1)
df_aux.drop(['DS_OBJETO'], axis=1, inplace=True) #podemos descartar o texto livre agora...
```

###### Agora podemos treinar o segundo modelo:

```python
X = df_aux.drop(["VL_LICITACAO"], axis=1, inplace=False)
#X = df_aux[["p1","p2","p3","p4","p5"]]
y = df_aux[["VL_LICITACAO"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
model = sm.OLS(y_train,X_train)
fitted = model.fit()
fitted.summary()
```

###### O resultado foi praticamente igual ao do último modelo (0.7 de acréscimo no R-squared):
ou seja, adicionar informação textual em forma de TF-IDF não ajuda um regressor linear a aproximar o VL_LICITACAO

###### Vamos tentar outra abordagem:
Vamos clusterizar os vetores TF-IDF das licitações e, ao invés de concatenar um PCA-like numérico ref. a licitação ao dataset, simplesmente iremos criar uma variável categórica indicando a qual grupo a licitação pertence de acordo com seu texto.

###### Quantos tipos de licitações existem?
Não sabemos, o número de clusters pode ser arbitrário mas o ideal é escolher aquele que causa menos separabilidade intra-cluster. Pode-se avaliar isso com o silhouette score

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import Normalizer
from sklearn.metrics import silhouette_score
svd = TruncatedSVD(n_components=500) #Infelizmente meu PC travou qndo tentei executar KMeans em 12k dimensões, então estou reduzindo...
normalizer = Normalizer(copy=False)
red = make_pipeline(svd,normalizer)
lic_embed_red_norm = red.fit_transform(lic_embed)
for n_c in range(2,16): #Vamos considerar que existam entre 2 e 15 tipos
    model = MiniBatchKMeans(n_clusters=n_c,random_state=20)
    cluster_labels = model.fit_predict(lic_embed_red_norm)
    print("For k={n_c}, avg silhouette score = {score}".format(n_c=n_c, score= silhouette_score(lic_embed_red_norm,cluster_labels)))
```

###### O silhouette score está bem baixo, o que é um mau sinal, mas vamos escolher k=15 já que este foi o maior.

```python
model = MiniBatchKMeans(n_clusters=15,random_state=20)
cluster_labels = model.fit_predict(lic_embed_red_norm)
```

###### Agora podemos "colar" essa nova coluna no dataset, junto com componentes do SVD de antes, já deixando ela one-encoded também:

```python
df_cluster = pd.DataFrame(cluster_labels, dtype=np.int32, columns=["CLUSTER"])
df_aux = pd.concat([df_aux_dum,  df_pcs, df_cluster], axis=1)
df_aux.drop(['DS_OBJETO'], axis=1, inplace=True) #podemos descartar o texto livre agora...
df_aux['CLUSTER'] = df_aux['CLUSTER'].apply(str)
df_aux = pd.get_dummies(data=df_aux,columns=['CLUSTER'],drop_first=True)
```

###### Nesse último modelo vamos usar uma regressão linear regularizada, que penaliza o uso de muitas variáveis

###### Vamos dividir o dataset em treino e teste:
não iremos utilizar as variaveis numericas p1, ...,p5 computadas antes, pois assumimos que a nova variavel categorica sumariza elas.

```python
X = df_aux.drop(["VL_LICITACAO","p1","p2","p3","p4","p5"], axis=1, inplace=False)
y = df_aux[["VL_LICITACAO"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

###### Agora instanciamos e executamos o modelo
cada variável é associada a um coeficiente beta.

```python
from sklearn.linear_model import LassoCV
alps = 10**np.linspace(-2,2,200)
ls_model = LassoCV(alphas=alps, cv=5)
ls_model.fit(X_train, np.ravel(y_train))

betas = list(zip(X_train.columns, ls_model.coef_))
#betas
```

###### Vamos ver como ficou a métrica R-squared em treino:

```python
ls_model.score(X_train,np.ravel(y_train))
```

###### O modelo permanece com performance baixa: 
vamos tentar outra abordagem. Ao invés de usar TF-IDF, vamos usar doc2vec, um modelo neural, para obter um vetor numérico para cada licitação. Cada licitação será representada por uma bag of words dos seus itens + sua descrição DS_OBJETO, para isso precisamos usar o dataset itens.csv.
Premissa: licitações semelhantes entre si, contém itens semelhantes

###### Agora vamos concatenar o dataset de itens ao nosso dataset 

```python
path_it = os.path.join(".",str(year),"item.csv")
df_it = pd.read_csv(path_it)
df_it= df_it[['CD_ORGAO','NR_LICITACAO','ANO_LICITACAO','CD_TIPO_MODALIDADE','NR_ITEM','DS_ITEM','QT_ITENS','VL_UNITARIO_ESTIMADO','VL_UNITARIO_HOMOLOGADO']]
mer = pd.merge(df_lic,df_it,on=['CD_ORGAO','NR_LICITACAO','ANO_LICITACAO','CD_TIPO_MODALIDADE'], how='left')
mer = mer[['CD_ORGAO','NR_LICITACAO','ANO_LICITACAO','CD_TIPO_MODALIDADE','DS_ITEM']]
mer['DS_ITEM'] = mer['DS_ITEM'].astype(str)
df_lic_it = mer.groupby(['CD_ORGAO','NR_LICITACAO','ANO_LICITACAO','CD_TIPO_MODALIDADE'])['DS_ITEM'].apply(' '.join).reset_index()
df_lic_it.head()
```

```python
df_lic_it = df_lic_it[~df_lic_it.DS_ITEM.str.match('nan')] #removemos licitações sem descrição de itens
df_lic_it.reset_index(inplace=True,drop=True)
```

###### Juntamos com as variáveis do dataset original.

```python
df_lic_it = pd.merge(df_lic_it, df_lic.drop(['NM_ORGAO', 'DT_ABERTURA'],axis=1,inplace=False), 
                     on=['CD_ORGAO', 'NR_LICITACAO', 'ANO_LICITACAO','CD_TIPO_MODALIDADE'], how='left')
```

###### Vamos remover as stopwords do campo DS_ITEM

```python
stop = stopwords.words('portuguese')
df_lic_it['DS_ITEM'] = df_lic_it['DS_ITEM'].apply(lambda x: ' '.join([w for w in x.split() if w not in (stop)])) 
```

###### Agora treinamos o modelo de fato:
tokenizamos as entradas primeiro
usamos um tamanho de embeddings=50 e 100 épocas de treinamento, dm=0 implica que usaremos bag of words, onde a ordem das palavras dentro do documento não importa.
ATENÇÃO: se o arquivo d2v.model já estiver na pasta, o treinamento não será executado, mas o modelo será carregado a partir do arquivo.

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
nltk.download('punkt')
data = list(df_lic_it['DS_ITEM'])
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[i]) for i, _d in enumerate(data)]

if os.path.isfile('d2v.model'):
    model = Doc2Vec.load('d2v.model')
    print('Modelo carregado')
else:
    model = Doc2Vec(vector_size=50,alpha=0.025, min_alpha=0.00025,min_count=2,dm =0, epochs=100)
    model.build_vocab(tagged_data)
    print("Treinando modelo...")
    model.train(tagged_data,total_examples=model.corpus_count,epochs=model.epochs)
    model.save("d2v.model") #salvar pra nao precisar treinar de novo
    print("Modelo salvo")
```

```python
model.delete_temporary_training_data(keep_doctags_vectors=True,keep_inference=True) #eliminando uso de mem desnecessário
```

###### Agora temos um vetor de 50 valores pra cada licitação:
assim como fizemos antes para o TF-IDF, vamos reduzir as dimensões desse vetor para 5, gerando 5 variáveis preditivas para o dataset.
Já concatenamos essas variaveis ao dataset e dropamos algumas colunas que nao serao utilizadas na regressão.

```python
from sklearn.decomposition import PCA
lic_embed = np.array([ model.docvecs[i] for i in range(0,df_lic_it.shape[0]) ])
lic_embed_red = PCA(n_components=5).fit_transform(lic_embed)
df_pcs = pd.DataFrame(lic_embed_red, dtype=np.float32, columns=["p1", "p2", "p3", "p4", "p5"])
df_LI_aux = pd.concat([df_lic_it,  df_pcs], axis=1).drop(['DS_ITEM','CD_ORGAO','NR_LICITACAO',
                                                              'ANO_LICITACAO'], axis=1)
```

###### Ainda preparando o dataset para o modelo de regressão:
encodamos as variáveis categóricas, removes observações cujo VL_LICITACAO é outlier e removemos a coluna TP_LICITACAO_NSA que tem alta correlação com outra variável, como vimos no heatmap.

```python
df_LI_aux_dum = pd.concat((df_LI_aux[["VL_LICITACAO", "p1", "p2", "p3","p4"]],
                      pd.get_dummies(data=df_LI_aux, columns=["MES_ABERTURA","CD_TIPO_MODALIDADE",
                                    "TP_COMISSAO","TP_LICITACAO", "TP_ORGAO",
                                    "TP_BENEFICIO_MICRO_EPP","TP_FORNECIMENTO",
                                    "BL_ORCAMENTO_SIGILOSO", "BL_PERMITE_CONSORCIO"], drop_first=True)), axis=1)
df_LI_aux_dum = df_LI_aux_dum.loc[:,~df_LI_aux_dum.columns.duplicated()]
q_low = df_LI_aux_dum["VL_LICITACAO"].quantile(0.10)
q_high = df_LI_aux_dum["VL_LICITACAO"].quantile(0.90)
df_LI_aux_dum = df_LI_aux_dum[(df_LI_aux_dum["VL_LICITACAO"] < q_high) & (df_LI_aux_dum["VL_LICITACAO"] > q_low)]
df_LI_aux_dum.drop(["TP_LICITACAO_NSA"], axis=1, inplace=True)
df_LI_aux_dum.reset_index(inplace=True,drop=True)
```

###### Observação:
depois de remover licitações cujas descrições dos items resultava em nan e remover observações outliers, acabamos com um dataset com 8725 observações, cerca de 2.5k a menos que o original.

```python
df_LI_aux_dum.shape
```

###### Agora vamos rodar o modelo OLS:
por simplificação, rodaremos ele no dataset inteiro. Apenas para avaliar a métrica R-squared

```python
X_train = df_LI_aux_dum.drop(['VL_LICITACAO','DS_OBJETO'], axis=1)
y_train = df_LI_aux_dum[['VL_LICITACAO']]
model = sm.OLS(y_train,X_train)
fitted = model.fit()
fitted.summary()
```
